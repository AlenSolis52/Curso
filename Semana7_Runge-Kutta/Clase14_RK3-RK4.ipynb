# <span style="font-family:Georgia; text-align:center;"> Métodos RK de **tres** etapas (orden 3)</span> 

## <span style="font-family:Georgia; text-align:center;">Esquema general y tabla de Butcher</span>
<span style="font-family:Georgia;">


Con tres etapas explícitas:
\begin{align*}
y_{n+1} &= y_n + h\,(b_1k_1 + b_2k_2 + b_3k_3),\\
k_1 &= f(t_n,\ y_n),\\
k_2 &= f\!\big(t_n+c_2 h,\ y_n + h\,a_{21} k_1\big),\\
k_3 &= f\!\big(t_n+c_3 h,\ y_n + h\,(a_{31} k_1 + a_{32} k_2)\big).\\[1mm]
\end{align*}

Tabla de Butcher:
$$
\begin{array}{c|ccc}
0\\
c_2 & a_{21}\\
c_3 & a_{31} & a_{32}\\ \hline
 & b_1 & b_2 & b_3
\end{array}
$$
con $c_2=a_{21}$ y $c_3=a_{31}+a_{32}$ en el caso explícito.

## <span style="font-family:Georgia; text-align:center;">Condiciones de orden 3 por Taylor</span>
<span style="font-family:Georgia;"> 

Repitiendo el análisis de Taylor (ahora hasta $O(h^4)$ para forzar coincidencia hasta $h^3$). La solución exacta satisface (expansión de Taylor alrededor de $(t_n,y_n)$):
$$
\begin{aligned}
y(t_n+h) &= y_n + h\,f + \frac{h^2}{2}\big(f_t + f_y f\big) 
+ \frac{h^3}{6}\Big(f_{tt} + 2 f_{ty} f + f_{yy} f^2 + f_y(f_t + f_y f)\Big)
+ O(h^4),
\end{aligned}
$$

### Expansión de $k_2= f\big(t_n+\Delta t,\ y_n + \Delta y\big)$

Aquí $\Delta t = c_2 h$, $\Delta y = h\,a_{21}k_1 = h\,a_{21} f$, por lo que
$$
\begin{aligned}
k_2 &= f + h\,(c_2 f_t + a_{21} f_y f)
+ \frac{h^2}{2}\Big(c_2^2 f_{tt} + 2 c_2 a_{21} f_{ty} f + a_{21}^2 f_{yy} f^2\Big)
+ O(h^3).
\end{aligned}
$$
Como usamos $c_2=a_{21}$, puede escribirse con $c_2$ únicamente si se desea.

### Expansión de $k_3= f\big(t_n+\Delta t,\ y_n + \Delta y\big)$

Aquí $\Delta t = c_3 h$ y
$$
\Delta y = h(a_{31}k_1 + a_{32}k_2)
= h\,(a_{31}+a_{32}) f + h^2 a_{32}\,(c_2 f_t + a_{21} f_y f) + O(h^3)
= h\,c_3 f + h^2 a_{32}\,(c_2 f_t + a_{21} f_y f) + O(h^3).
$$
Entonces
$$
\begin{aligned}
k_3 &= f + h\,(c_3 f_t + c_3 f_y f)
\\ &\quad
+ \frac{h^2}{2}\Big(c_3^2 f_{tt} + 2 c_3^2 f_{ty} f + c_3^2 f_{yy} f^2\Big)
+ h^2\Big(a_{32} c_2\, f_y f_t + a_{32} a_{21}\, f_y(f_y f)\Big)
+ O(h^3).
\end{aligned}
$$


<span style="font-family:Georgia;"> 

Sustituimos las expansiones en
$$
y_{n+1} = y_n + h(b_1 k_1 + b_2 k_2 + b_3 k_3)
$$
y coleccionamos por potencias de $h$ y por las "bases" 
$\{f,\ f_t,\ f_y f,\ f_{tt},\ f_{ty} f,\ f_{yy} f^2,\ f_y f_t,\ f_y(f_y f)\}$.

### Orden $h^1$
$$
h:\quad b_1 + b_2 + b_3 = 1.
$$

### Orden $h^2$
Los coeficientes de $f_t$ y $f_y f$ deben igualar los de Taylor ($1/2$ cada uno):
$$
\begin{aligned}
h^2\!\!:&\quad b_2 c_2 + b_3 c_3 = \tfrac{1}{2},\\
        &\quad b_2 a_{21} + b_3 (a_{31}+a_{32}) = \tfrac{1}{2}.
\end{aligned}
$$
Con la consistencia $c_2=a_{21}$ y $c_3=a_{31}+a_{32}$, **ambas ecuaciones son la misma**:
$$
b_2 c_2 + b_3 c_3 = \tfrac{1}{2}.
$$

### Orden $h^3$
Al comparar con la serie exacta se obtienen cuatro ecuaciones independientes:
$$
\begin{aligned}
&\text{(i) } b_2 c_2^2 + b_3 c_3^2 = \tfrac{1}{3} \qquad &&\text{(coef. de $f_{tt}$)},\\[3pt]
&\text{(ii) } b_2 c_2^2 + b_3 c_3^2 = \tfrac{1}{3} \qquad &&\text{(coef. de $f_{ty}f$; coincide con (i) por $c=A\mathbf{1}$)},\\[3pt]
&\text{(iii) } b_2 c_2^2 + b_3 c_3^2 = \tfrac{1}{3} \qquad &&\text{(coef. de $f_{yy}f^2$; idem)},\\[3pt]
&\text{(iv) } b_3\,a_{32}\,c_2 = \tfrac{1}{6} \qquad &&\text{(coef. de $f_y f_t$)},\\[3pt]
&\text{(v) } b_3\,a_{32}\,a_{21} = \tfrac{1}{6} \qquad &&\text{(coef. de $f_y (f_y f)$)}.
\end{aligned}
$$

Bajo $c_2=a_{21}$ y $c_3=a_{31}+a_{32}$, las condiciones (ii) y (iii) **coinciden con** (i),
y (iv)–(v) **coinciden entre sí**. Por tanto, la **familia mínima** de condiciones de orden hasta 3
para RK explícito con $c=A\mathbf{1}$ se reduce a (forma vectorial estándar):
$$
\boxed{
\begin{aligned}
&\text{(1) } b^\top \mathbf{1} = 1,\\
&\text{(2) } b^\top c = \tfrac{1}{2},\\
&\text{(3) } b^\top (c \odot c) = \tfrac{1}{3},\\
&\text{(4) } b^\top A c = \tfrac{1}{6}.
\end{aligned}
}
$$
Aquí $A=(a_{ij})$, $b=(b_1,b_2,b_3)^\top$, $c=(c_1,c_2,c_3)^\top$ y $\odot$ es el producto elemento a elemento.
Para nuestro $A$ triangular inferior,
$$
b^\top A c = b_3\,a_{32}\,c_2.
$$
Analogamente 
$$
\boxed{
\begin{aligned}
&\text{(1)}\quad b_1 + b_2 + b_3 = 1,\\[6pt]
&\text{(2)}\quad b_2 c_2 + b_3 c_3 = \tfrac{1}{2},\\[6pt]
&\text{(3)}\quad b_2 c_2^2 + b_3 c_3^2 = \tfrac{1}{3},\\[6pt]
&\text{(4)}\quad b_3 a_{32} c_2 = \tfrac{1}{6}.
\end{aligned}}
$$

## <span style="font-family:Georgia; text-align:center;">Una solución particular: RK3 clásico (3 etapas)</span>
<span style="font-family:Georgia;"> 

Una elección estándar que satisface (1)–(4) es
$$
\begin{aligned}
c_2 &= \tfrac{1}{2},\quad &c_3 &= 1,\\
a_{21} &= \tfrac{1}{2},\quad &a_{31} &= -1,\quad &a_{32}&=2,\\
b_1 &= \tfrac{1}{6},\quad &b_2 &= \tfrac{4}{6},\quad &b_3&=\tfrac{1}{6}.
\end{aligned}
$$

Tabla de Butcher:
$$
\begin{array}{c|ccc}
0   &      &   &   \\
1/2 & 1/2  &   &   \\
1   & -1   & 2 &   \\
\hline
    & 1/6  & 4/6 & 1/6
\end{array}
$$

El método resultante es
$$
\begin{aligned}
y_{n+1} &= y_n + \tfrac{h}{6}\,(k_1 + 4k_2 + k_3),\\
k_1 &= f(t_n,y_n),\\
k_2 &= f\!\Big(t_n+\tfrac{h}{2},\; y_n+\tfrac{h}{2}\,k_1\Big),\\
k_3 &= f\!\big(t_n+h,\; y_n - h\,k_1 + 2h\,k_2\big).\\
\end{aligned}
$$
que es el **RK3 clásico**.

## <span style="font-family:Georgia; text-align:center;">Métodos RK de **cuatro** etapas (orden 4)</span>
<span style="font-family:Georgia;"> 

Con cuatro etapas explícitas :
\begin{align*}
y_{n+1} &= y_n + h\,(b_1k_1+b_2k_2+b_3k_3+b_4k_4),\\
k_1 &= f(t_n, y_n),\\
k_2 &= f\!\big(t_n+c_2 h,\ y_n+h\,a_{21}k_1\big),\\
k_3 &= f\!\big(t_n+c_3 h,\ y_n+h\,(a_{31}k_1+a_{32}k_2)\big),\\
k_4 &= f\!\big(t_n+c_4 h,\ y_n+h\,(a_{41}k_1+a_{42}k_2+a_{43}k_3)\big).\\[1mm]
\end{align*}
donde  $c=A\mathbf{1}$ con
$c_2=a_{21}$, $c_3=a_{31}+a_{32}$, $c_4=a_{41}+a_{42}+a_{43}$.

### Condiciones de orden 4 (vía Taylor hasta $O(h^5)$)

Las **condiciones necesarias y suficientes** para **orden 4** son (forma compacta estándar):
$$
\boxed{
\begin{aligned}
&\text{(1)}\quad b^\top \mathbf{1} = 1,\\
&\text{(2)}\quad b^\top c = \tfrac12,\\
&\text{(3)}\quad b^\top (c \odot c) = \tfrac13,\\
&\text{(4)}\quad b^\top (c \odot c\odot c) = \tfrac14,\\
&\text{(5)}\quad b^\top A c = \tfrac16,\\
&\text{(6)}\quad b^\top A (c \odot c) = \tfrac{1}{12},\\
&\text{(7)}\quad b^\top\big( (A c)\odot c \big) = \tfrac18,\\
&\text{(8)}\quad b^\top A^2 c = \tfrac{1}{24}.
\end{aligned}}
$$
Analogamente
$$
\boxed{
\begin{aligned}
&\text{(1)}\quad b_1 + b_2 + b_3 + b_4 = 1,\\[6pt]
&\text{(2)}\quad b_2 c_2 + b_3 c_3 + b_4 c_4 = \tfrac{1}{2},\\[6pt]
&\text{(3)}\quad b_2 c_2^2 + b_3 c_3^2 + b_4 c_4^2 = \tfrac{1}{3},\\[6pt]
&\text{(4)}\quad b_2 c_2^3 + b_3 c_3^3 + b_4 c_4^3 = \tfrac{1}{4},\\[6pt]
&\text{(5)}\quad b_3 a_{32} c_2 + b_4 a_{42} c_2 + b_4 a_{43} c_3 = \tfrac{1}{6},\\[6pt]
&\text{(6)}\quad b_3 a_{32} c_2^2 + b_4 a_{42} c_2^2 + b_4 a_{43} c_3^2 = \tfrac{1}{12},\\[6pt]
&\text{(7)}\quad b_3 c_3 a_{32} c_2 + b_4 c_4 a_{42} c_2 + b_4 c_4 a_{43} c_3 = \tfrac{1}{8},\\[6pt]
&\text{(8)}\quad b_4 a_{43} a_{32} c_2 = \tfrac{1}{24}.
\end{aligned}}
$$
**Grados de libertad.** Un RK explícito de 4 etapas tiene $6$ coeficientes subdiagonales en $A$ y $4$ pesos $b_i$; con $c=A\mathbf{e}$ hay **10** incógnitas frente a **8** ecuaciones ⇒ quedan **2 parámetros libres** (familia bidimensional).

### Dos métodos clásicos de orden 4

**(a) Kutta clásico**
$$
\begin{array}{c|cccc}
0\\
\frac12 & \frac12\\
\frac12 & 0 & \frac12\\
1 & 0 & 0 & 1\\ \hline
 & \frac16 & \frac13 & \frac13 & \frac16
\end{array}
$$

**(b) Regla de $3/8$**
$$
\begin{array}{c|cccc}
0\\
\frac13 & \frac13\\
\frac23 & -\frac13 & 1\\
1 & 1 & -1 & 1\\ \hline
 & \frac18 & \frac{3}{8} & \frac{3}{8} & \frac18
\end{array}
$$

Ambos satisfacen (1)–(8).

# <span style="font-family:Georgia; text-align:center;">Ejercicio</span>
<span style="font-family:Georgia;"> 

La reacción química irreversible en la que **dos moléculas de dicromato de potasio sólido** ($K_2Cr_2O_7$),  **dos moléculas de agua** ($H_2O$) y **tres átomos de azufre sólido** ($S$) se combinan para producir **cuatro moléculas de hidróxido de potasio** ($KOH$), **dos moléculas de óxido de cromo** ($Cr_2O_3$) y **tres moléculas de dióxido de azufre** ($SO_2$), puede representarse mediante la ecuación estequiométrica:

$$
2K_2Cr_2O_7 + 2H_2O + 3S \;\longrightarrow\; 4KOH + 2Cr_2O_3 + 3SO_2.
$$

Sean:
- $n_1$: número inicial de moléculas de $K_2Cr_2O_7$,
- $n_2$: número inicial de moléculas de $H_2O$,
- $n_3$: número inicial de átomos de $S$,
- $x(t)$ la cantidad de $KOH$ formada en el tiempo $t$.

De la ecuación estequiométrica podemos observar:

- Por cada **2 moléculas de $K_2Cr_2O_7$** consumidas se producen **4 de $KOH$**,  
  es decir, cada unidad de $KOH$ corresponde a $\tfrac{1}{2}$ de $K_2Cr_2O_7$.
- Por cada **2 moléculas de $H_2O$** se generan **4 de $KOH$**,  
  luego cada unidad de $KOH$ corresponde a $\tfrac{1}{2}$ de $H_2O$.
- Por cada **3 átomos de $S$** se forman **4 de $KOH$**,  
  lo que implica que cada unidad de $KOH$ equivale a $\tfrac{3}{4}$ de $S$.

Por tanto, después de haberse producido $x$ unidades de $KOH$, las cantidades restantes son:

$$
\begin{aligned}
\text{Dicromato:}&\quad n_1 - \tfrac{x}{2},\\[4pt]
\text{Agua:}&\quad n_2 - \tfrac{x}{2},\\[4pt]
\text{Azufre:}&\quad n_3 - \tfrac{3x}{4}.
\end{aligned}
$$


Según la **ley de acción de masas**, la velocidad de una reacción es proporcional al producto de las concentraciones de los reactivos, cada una elevada a la potencia correspondiente a su coeficiente estequiométrico:

- El $K_2Cr_2O_7$ participa con coeficiente **2** $\;\Rightarrow\;$ potencia cuadrática.  
- El $H_2O$ también con coeficiente **2** $\;\Rightarrow\;$ potencia cuadrática.  
- El azufre $S$ interviene con coeficiente **3** $\;\Rightarrow\;$ potencia cúbica.

De esta forma, la evolución temporal de $KOH$ está dada por:

$$
\frac{dx}{dt} \;=\; k \left(n_1 - \tfrac{x}{2}\right)^2
                    \left(n_2 - \tfrac{x}{2}\right)^2
                    \left(n_3 - \tfrac{3x}{4}\right)^3,
$$

donde $k$ es la constante de velocidad de la reacción.

Si
$$
k = 6.22 \times 10^{-19}, \quad
n_1 = n_2 = 2\times 10^3, \quad
n_3 = 3\times 10^3,
$$

utiliza el **método de Runge–Kutta de orden cuatro** para estimar cuántas unidades de $KOH$ se han formado después de $0.2$ segundos.


```python
# Solve the ODE from the problem using classical RK4 with step h=0.01 up to t=0.2
import numpy as np

k = 6.22e-19
n1 = 2e3
n2 = 2e3
n3 = 3e3

def f(t, x):
    return k * (n1 - x/2.0)**2 * (n2 - x/2.0)**2 * (n3 - 3.0*x/4.0)**3

t0 = 0.0
x0 = 0.0   # initially no KOH formed
T  = 0.2
h  = 0.01  # step

def rk4(f, t0, x0, h, T):
    t = t0
    x = x0
    while t < T - 1e-15:
        if t + h > T:  # last partial step if needed
            h = T - t
        k1 = f(t, x)
        k2 = f(t + 0.5*h, x + 0.5*h*k1)
        k3 = f(t + 0.5*h, x + 0.5*h*k2)
        k4 = f(t + h, x + h*k3)
        x = x + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)
        t = t + h
    return x

x_T = rk4(f, t0, x0, h, T)
x_T
```
