# <span style="font-family:Georgia; text-align:center;">Metodos en 2 etapas</span> 

## <span style="font-family:Georgia; text-align:center;">Métodos de orden 2</span>
<span style="font-family:Georgia;">

El método del punto medio está definido por definido por:
$$
\hat y_0 = y_0,\qquad
\hat y_{i+1}=\hat y_i + h\, f\!\left(t_i+\frac{h}{2},\, \hat y_i+\frac{h}{2}\, f(t_i,\hat{y_i})\right),\qquad i\in\{0,\ldots,N\},
$$

El método del trapecio está definido por definido por:
$$
\hat y_0 = y_0,\qquad
\hat{y}_{i+1} \;=\; \hat{y}_i \;+\; \frac{h}{2}\Big[f(t_i,\hat{y}_i)+f(t_{i}+h,\hat y_i+f(t_i,\hat{y}_i))\Big],\qquad i\in\{0,\ldots,N\},
$$

## <span style="font-family:Georgia; text-align:center;">Esquema RK de 2 etapas</span>
<span style="font-family:Georgia;">

Partimos de la forma
$$
y_{n+1} \;=\; y_n \;+\; h\big(b_1\,k_1 + b_2\,k_2\big),\qquad b_1,b_2\in\mathbb{R},
$$
donde, por ahora,
$$
k_1=f(t_n,y_n),\qquad k_2\ \text{es otra evaluación de}\ f\ \text{(por definir)}.
$$
- En el **punto medio**, los pesos son $(b_1,b_2)=(0,1)$.
- En el **trapecio**, los pesos son $(b_1,b_2)=(\tfrac12,\tfrac12)$.

## <span style="font-family:Georgia; text-align:center;">Forma general de RK **en 2 etapas** y **tabla de Butcher**</span>
<span style="font-family:Georgia;">

Un método RK **explícito** de 2 etapas tiene la forma
\begin{align*}
y_{n+1} &= y_n + h\big(b_1\,k_1 + b_2\,k_2\big),\\
k_1 &= f(t_n,\ y_n),\\
k_2 &= f\!\big(t_n + c_2h,\ y_n + h\,a_{21}\,k_1\big).\\[1mm]
\end{align*}


Su **tabla de Butcher** es
$$
\begin{array}{c|cc}
0 \\
c_2 & a_{21}\\ \hline
 & b_1 & b_2
\end{array}
$$

- Los **$c_i$** (aquí solo $c_2$) son los **tiempos internos**: se evalúa $f$ en $t_n+c_i h$.  
- Los **$b_i$** son los **pesos** con que se promedian las pendientes $k_i$ en la actualización final.  
- Los **$a_{ij}$** indican **cómo se combinan** las pendientes anteriores al formar los argumentos de $f$ en etapas posteriores.  
- En métodos explícitos, **$c_i=\sum_{j<i} a_{ij}$** (para $i\ge 2$). En particular, aquí $c_2=a_{21}$.

**Para los métodos recordados:**
- Punto medio: $c_2=a_{21}=\tfrac12$, $b=(0,1)$.
- Trapecio (Heun): $c_2=a_{21}=1$, $b=(\tfrac12,\tfrac12)$.


## <span style="font-family:Georgia; text-align:center;">Restricciones a las constantes (RK2) por expansión de Taylor</span>
<span style="font-family:Georgia;">

Usando Tayloe en la solución exacta y  abreviando $f_n:=f(t_n,y_n)$
$$
y(t_{n+1}) = y(t_{n}+h) = y_n + h\,f_n + \frac{h^2}{2}\, \frac{d}{dt}f_n + \mathcal{O}(h^3),
$$
Entonces
$$
\begin{aligned}
y(t_n+h) &= y_n + h\,f + \frac{h^2}{2}\big(f_t + f_y f\big) 
+ O(h^3),
\end{aligned}
$$

Usando Taylor en el esquema de 2 etapas tenemos:

- Primero $k_1=f_n$.
- Para $k_2$ usamos Taylor multivariable en $(t,y)$ con
$\Delta t=c_2h$, $\Delta y = h\,a_{21}\,k_1 = h\,a_{21}\,f_n$
\begin{align*}
k_2 \;=\; f(t_n+\Delta t,\ y_n+\Delta y)
&= f_n + \Delta t\,f_t + f_y\,\Delta y \\
&\quad+ \frac12\,\Delta t^2\,f_{tt} + \,f_{ty}\,\Delta t\,\Delta y + \frac12\,f_{yy}\Delta y^2 + \mathcal{O}(h^3)\\
&= f_n + h\big(c_2 f_t + a_{21} f_y f_n\big)\\
&\quad+ \frac{h^2}{2}\Big(c_2^2 f_{tt} + 2 c_2 a_{21} f_{ty} f_n + a_{21}^2 f_{yy}f_n^2\Big) + \mathcal{O}(h^3).
\end{align*}

El método produce
\begin{align*}
y_{n+1}
&= y_n + h\big(b_1 k_1 + b_2 k_2\big)\\
&= y_n + h(b_1+b_2) f_n \\
&\quad + h^2\,b_2\big(c_2 f_t + a_{21} f_y f_n\big) \\
&\quad  + \mathcal{O}(h^3).
\end{align*}

Para que el método sea **de orden 2**, debe coincidir con la solución exacta hasta $h^2$. Comparando coeficientes de $f_n$ y de $ \frac{d}{dt}f_n=f_t+f_y f_n$ se obtiene:
$$
\boxed{\,b_1+b_2=1,\qquad b_2 c_2=\tfrac12,\qquad b_2 a_{21}=\tfrac12\,}.
$$
En métodos RK **explícitos** se cumple $c_2=a_{21}$, de modo que las dos últimas igualdades **coinciden** y basta exigir
$$
\boxed{\,b_1+b_2=1,\qquad b_2 c_2=\tfrac12\,}.
$$

## <span style="font-family:Georgia; text-align:center;">Método de dos etapas basado en cuadratura con parámetro</span>
<span style="font-family:Georgia;"> 

A partir de fórmulas de cuadratura es posible construir esquemas de dos etapas que alcanzan orden dos.  
La idea consiste en realizar un avance intermedio de tamaño $\theta h$, y luego combinar la información de la derivada al inicio del paso y en este punto intermedio usando una fórmula de integración de segundo orden.  

En particular, se utiliza la aproximación

$$
\int_0^1 \varphi(x)\,dx \;\approx\; 
\Big(1 - \tfrac{1}{2\theta}\Big)\varphi(0) \;+\; \tfrac{1}{2\theta}\,\varphi(\theta),
$$

lo que conduce al siguiente esquema de dos etapas:

$$
\begin{aligned}
Y &= y_{n-1} + \theta h f(x_{n-1}, y_{n-1}), \\
y_{n} &= y_{n-1} + \Big(1 - \tfrac{1}{2\theta}\Big) h f(x_{n-1}, y_{n-1})
              + \tfrac{1}{2\theta} h f(x_{n-1} + \theta h,\; Y).
\end{aligned}
$$
### Tabla de Butcher

La representación en tabla de Butcher es

$$
\begin{array}{c|cc}
0 \\
\theta & \theta \\ \hline
 & 1 - \tfrac{1}{2\theta} & \tfrac{1}{2\theta}
\end{array}
$$

Este método depende de un único parámetro $\theta$.  
- Para $\theta=\tfrac{1}{2}$ se obtiene el **método del punto medio**.  
- Para $\theta=1$ se recupera el **método de trapecio**.
