# <span style="font-family:Georgia; text-align:center;"> Problema 1</span> 
<span style="font-family:Georgia;">

Sea $M>1$. Resuelve la ecuación diferencial:

$$
\begin{cases}
y'=\min\{y^2,\, M\},\\
y(0)=1.
\end{cases}
$$

¿La solución es única y global?


<span style="font-family:Georgia;">

## Solución

Definamos $f(t,y)=\min\{y^2,M\}$. La función $f$ es continua (por continuidad de $\min$). Además,
$$
\frac{\partial f}{\partial y}(t,y)=
\begin{cases}
2y, & |y|<\sqrt{M},\\[2mm]
0, & |y|\ge \sqrt{M},
\end{cases}
$$
por lo que $\big|\partial f/\partial y\big|\le 2\sqrt{M}$ para todo $y\in\mathbb{R}$. Luego $f$ es **Lipschitz global en $y$** en $[a,b]\times\mathbb{R}$, y por el Teorema de Existencia y Unicidad el PVI tiene **solución única y global** en cualquier intervalo $[a,b]$.

Mientras $|y(t)|<M$ se cumple $y' = y^2$, cuya solución con $y(0)=1$ es
$$
y(t)=\frac{1}{1-t}.
$$
Esta rama es válida hasta el tiempo $T$ en que $y(T)=\sqrt{M}$, es decir
$$
T=1-\frac{1}{\sqrt{M}}\in(0,1),\qquad y(T)=\sqrt{M}.
$$
A partir de $t\ge T$, la ecuación se satura y queda $y'=M$ con condición $y(T)=\sqrt{M}$, de donde
$$
y(t)=\sqrt{M}+\sqrt{M}(t-T).
$$

**Finalmete, tenemos la solución**
$$
y(t)=
\begin{cases}
\dfrac{1}{1-t}, & t<T,\\[3mm]
\sqrt{M}+\sqrt{M}(t-T), & t\ge T,
\end{cases}
\qquad T=1-\dfrac{1}{\sqrt{M}}.
$$

# <span style="font-family:Georgia; text-align:center;"> Problema 2</span> 
<span style="font-family:Georgia;">

Considera la ecuación diferencial:
$$
\begin{cases}
y' = e^{-y}\,\sqrt{e^{y}-1},\\
y(0)=0.
\end{cases}
$$

¿La solución de este problema es única? En caso negativo, determina (al menos) dos soluciones distintas.


<span style="font-family:Georgia;">

## Solución


Tomemos el cambio de variable $x(t)=e^{y(t)}$. Como $x' = e^{y}y'$, al sustituir obtenemos
$$
\frac{x'}{x} \;=\; \frac{1}{x}\sqrt{x-1}\quad\Longrightarrow\quad
x'=\sqrt{x-1},\qquad x(0)=e^{y(0)}=1.
$$
La función $g(x)=\sqrt{x-1}$ **no es Lipschitz** en $x=1$, de modo que el Teorema de Existencia y Unicidad no garantiza unicidad.

# <span style="font-family:Georgia; text-align:center;"> Problema 3</span> 
<span style="font-family:Georgia;">

Considere el PVI
$$
y' = e^{y}+e^{t},\qquad y(0)=0.
$$

1. Encuentra su solución.


<span style="font-family:Georgia;">

## Solución
Consideremos el cambio de avriable $z=e^{-y}$. Con $z=e^{-y}$ se tiene $y'=-\dfrac{z'}{z}$. La ecuación queda
$$
-\frac{z'}{z}=\frac{1}{z}+e^{t}\quad\Longrightarrow\quad
z'+e^{t}z=-1,\qquad z(0)=e^{-y(0)}=1.
$$
Es una EDO **lineal**. Tomando el factor integrante
$$
\mu(t)=\exp\!\Big(\int e^{t}\,dt\Big)=e^{\,e^{t}},
$$
obtenemos
$$
(\mu z)'=-\mu \;\;\Longrightarrow\;\; 
\mu(t)z(t)-\mu(0)z(0)=-\int_{0}^{t}\mu(s)\,ds.
$$
Como $\mu(0)=e$ y $z(0)=1$,
$$
z(t)=e^{-e^{t}}\!\left[e-\int_{0}^{t} e^{\,e^{s}}\,ds\right],
$$
