
# Bifurcaciones en EDOs unidimensionales



## Flujos en la recta: EDOs unidimensionales

Estudiaremos ecuaciones de la forma
$$
x' = f(x;\,\mu),
$$
donde $x \in \mathbb{R}$, $f:\mathbb{R}\times\mathbb{R}\to\mathbb{R}$ es suficientemente suave y $\mu$ es un parámetro. Para un valor fijo de $\mu$, la ecuación genera un campo de velocidades en la recta.

Bajo condiciones estándar de Lipschitz local en $x$, el **teorema de existencia y unicidad** garantiza que, dado un estado inicial $x(0)=x_0$, existe una solución única $x(t)$ definida en un intervalo alrededor de $t=0$.

## La gráfica $x$ vs $x'$ como herramienta de estabilidad

Los **puntos de equilibrio** (o fijos) satisfacen $f(x^*;\,\mu)=0$.  
Para estudiar su estabilidad usamos la **linealización**, que se justifica aplicando el **teorema de Taylor** a $f(x;\,\mu)$ alrededor del punto de equilibrio $x^*$:

$$
f(x;\,\mu) \;=\; f(x^*;\,\mu) + f'(x^*;\,\mu)\,(x-x^*) + R_2(x),
$$

donde $R_2(x)$ es el residuo de orden cuadrático o superior en $(x-x^*)$.  
Como $f(x^*;\,\mu)=0$, la ecuación diferencial se aproxima, cerca de $x^*$, por

$$
x' = f(x;\,\mu) \;\approx\; f'(x^*;\,\mu)\,(x-x^*).
$$

- Si $f'(x^*;\,\mu) < 0$, entonces $x^*$ es **estable** (atractor).
- Si $f'(x^*;\,\mu) > 0$, entonces $x^*$ es **inestable** (repulsor).
- Si $f'(x^*;\,\mu) = 0$, el equilibrio es **no hiperbólico**; su estabilidad depende de términos no lineales.

La **gráfica $x$ vs $x'$** (esto es, la gráfica de $x \mapsto f(x;\,\mu)$) permite:
1. Localizar equilibrios como intersecciones con el eje $x'$ ($f(x;\,\mu)=0$).
2. Inferir estabilidad observando el signo de la pendiente en la intersección: pendiente negativa $\Rightarrow$ estable; pendiente positiva $\Rightarrow$ inestable.


### Ejemplo: la ecuación logística en dimensión 1

Consideremos la ecuación diferencial

$$
x' = r\,x\,(1 - x), \quad r > 0,
$$

donde $x(t)$ representa la población normalizada (entre $0$ y $1$) y $r$ es la tasa de crecimiento.

**Equilibrios:**
- $x^* = 0$,
- $x^* = 1$.

Aplicamos el **teorema de Taylor** centrado en cada equilibrio:

1. Para $x^*=0$:
$$
f(x) = r x(1-x) = f(0) + f'(0)(x-0) + R_2(x),
$$
donde $f(0)=0$ y $f'(0)=r >0$.  
Por tanto, $x=0$ es **inestable**.

2. Para $x^*=1$:
$$
f(x) = r x(1-x) = f(1) + f'(1)(x-1) + R_2(x),
$$
donde $f(1)=0$ y $f'(1) = r(1-2) = -r < 0$.  
Por tanto, $x=1$ es **estable**.

**Interpretación:**  
- Si la población parte cerca de $0$, crecerá hacia el equilibrio $1$.  
- Si la población parte cerca de $1$, se mantendrá estable en ese nivel.


## ¿Qué es una bifurcación?

Una **bifurcación** ocurre en un valor crítico $\mu=\mu_c$ cuando un cambio pequeño del parámetro produce un **cambio cualitativo** en la dinámica: cambia el número de puntos de equilibrio y/o su estabilidad (es decir, cambia la topologia del retrato de fases).

En dimensión $1$, las bifurcaciones genéricas son:
- **Silla–nodo**.
- **Transcrítica**.
- **Tridente supercrítica** (*pitchfork* supercrítica).
- **Tridente subcrítica** (*pitchfork* subcrítica).



## Formas normales (modelos canónicos)



### Silla–nodo: $x'=\mu - x^2$

**Equilibrios**: $x = \pm\sqrt{\mu}$ si $\mu>0$; no hay equilibrios si $\mu<0$; en $\mu=0$ aparece un equilibrio **no hiperbólico** en $x=0$.  
**Estabilidad**: $f'(x) = -2x$. Para $\mu>0$, $x=+\sqrt{\mu}$ es **estable** y $x=-\sqrt{\mu}$ es **inestable**.


```python
interact(graficar_silla_nodo, mu=FloatSlider(min=-2.0, max=2.0, step=0.01, value=0.0, description='μ'))
```


### Transcrítica: $x'=\mu x - x^2$

**Equilibrios**: $x=0$ y $x=\mu$.  
**Estabilidad**: $f'(x)=\mu - 2x$. En $\mu<0$: $x=0$ es **estable** y $x=\mu$ es **inestable**. En $\mu>0$ se invierten: $x=0$ es **inestable** y $x=\mu$ es **estable**. En $\mu=0$ colisionan en un equilibrio **no hiperbólico**.


```python
interact(graficar_transcritica, mu=FloatSlider(min=-2.0, max=2.0, step=0.01, value=0.0, description='μ'))
```


### Tridente supercrítica (*pitchfork* supercrítica): $x'=\mu x - x^3$

**Equilibrios**: $x=0$ para todo $\mu$; además $x=\pm\sqrt{\mu}$ si $\mu>0$.  
**Estabilidad**: $f'(x)=\mu - 3x^2$.  
- Para $\mu<0$: $x=0$ es **estable**.  
- Para $\mu>0$: $x=0$ es **inestable** y aparecen dos equilibrios **estables** en $x=\pm\sqrt{\mu}$.


```python
interact(graficar_tridente_supercritica, mu=FloatSlider(min=-2.0, max=2.0, step=0.01, value=0.0, description='μ'))
```


### Tridente subcrítica (*pitchfork* subcrítica): $x'=\mu x + x^3$

**Equilibrios**: $x=0$ para todo $\mu$; además $x=\pm\sqrt{-\mu}$ si $\mu<0$.  
**Estabilidad**: $f'(x)=\mu + 3x^2$.  
- Para $\mu<0$: $x=0$ es **estable**, mientras que $x=\pm\sqrt{-\mu}$ son **inestables**.  
- Para $\mu>0$: $x=0$ es **inestable** y no hay otros equilibrios.


```python
interact(graficar_tridente_subcritica, mu=FloatSlider(min=-2.0, max=2.0, step=0.01, value=0.0, description='μ'))
```
