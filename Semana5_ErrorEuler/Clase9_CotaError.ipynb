
# <span style="font-family:Georgia; text-align:center;">Cotas de error para el método de Euler</span> 

## <span style="font-family:Georgia; text-align:center;">Proposición 1</span> 
<span style="font-family:Georgia;">

Si $x\ge -1$ y $m\in\mathbb{N}$, entonces
$$
0 \le (1+x)^m \le e^{mx}.
$$

**Prueba.** Por el Teorema de Taylor para $f(x)=e^{x}$ alrededor de $0$ con resto de segundo orden,
$$
e^{x} \;=\; 1 + x + \tfrac{1}{2}x^2 e^{\xi},
$$
para algún $\xi$ entre $0$ y $x$. De aquí se deduce $1+x\le e^{x}$ y $1+x\ge 0$ si $x\ge -1$. Elevando a la potencia $m$ (válido porque $1+x\ge 0$), obtenemos
$$
0 \le (1+x)^m \le (e^x)^m = e^{mx}.
$$
$\square$


## <span style="font-family:Georgia; text-align:center;">Proposición 2: Desigualdad de Grönwall discreta</span> 
<span style="font-family:Georgia;">

Sean $s,t>0$ y una sucesión $(a_i)_{i=0}^{k}$ con $a_0\ge -t/s$ que satisface
$$
a_{i+1}\le (1+s)\,a_i + t,\qquad i=0,1,\dots,k-1.
$$
Entonces
$$
a_{i+1}\;\le\; e^{(i+1)s}\!\left(a_0+\frac{t}{s}\right)-\frac{t}{s}.
$$

**Prueba.** Iteramos la desigualdad:
\begin{align*}
a_{i+1} &\le (1+s)a_i + t
\;\\&\le\; (1+s)^2 a_{i-1} + \big[1+(1+s)\big]t\\ &\quad \vdots\\
&\le\; (1+s)^{i+1} a_0 + \big[1+(1+s)+\cdots+(1+s)^i\big]t.
\end{align*}
La suma geométrica vale 
$$
1+(1+s)+\cdots+(1+s)^i=\frac{(1+s)^{i+1}-1}{(1+s)-1}=\frac{1-(1+s)^{i+1}}{1-(1+s)}=\frac{1}{s}[(1+s)^{i+1}-1], 
$$
así que
$$
a_{i+1}\;\le\; (1+s)^{i+1}\!\left(a_0+\frac{t}{s}\right)-\frac{t}{s}.
$$
Aplicando la **Proposición 1** con $x=s$ (nótese $s>-1$) se obtiene $(1+s)^{i+1}\le e^{(i+1)s}$ y queda
$$
a_{i+1}\;\le\; e^{(i+1)s}\!\left(a_0+\frac{t}{s}\right)-\frac{t}{s}.
$$
$\square$


## <span style="font-family:Georgia; text-align:center;">Teorema: Cota global de error de Euler</span> 
<span style="font-family:Georgia;">

Suponga que $f$ es continua y satisface la condición de Lipschitz en $y$ con constante $L$ en
$$
D=\{(t,y)\mid a\le t\le b,\; -\infty<y<\infty\},
$$
y que existe $M>0$ tal que $|y''(t)|\le M$ para toda $t\in[a,b]$, donde $y$ es la **única** solución del PVI
$$
y'(t)=f(t,y(t)),\qquad a\le t\le b,\qquad y(a)=\alpha.
$$
Sea $\hat{y}_0,\dots,\hat{y}_N$ la sucesión de Euler con paso $h=\frac{b-a}{N}$. Entonces, para cada $i=0,1,\dots,N,$
$$
\boxed{\;|y(t_i)-\hat{y}_i|\;\le\;\frac{hM}{2L}\,\big(e^{L(t_i-a)}-1\big)\;}
$$
donde $t_i=a+ih$.

**Prueba.** Por Taylor en $[t_i,t_{i+1}]$ existe $\xi_i\in(t_i,t_{i+1})$ con
$$
y(t_{i+1})=y(t_i)+h\,y'(t_i)+\tfrac{h^2}{2}\,y''(\xi_i)
\;=\; y(t_i)+h\,f(t_i,y(t_i))+\tfrac{h^2}{2}\,y''(\xi_i).
$$
El método de Euler satisface
$$
\hat{y}_{i+1}=\hat{y}_i+h\,f(t_i,\hat{y}_i).
$$
Restando ambas expresiones y usando $y_i:=y(t_i)$,
$$
y_{i+1}-\hat{y}_{i+1}=(y_i-\hat{y}_i)+h\big(f(t_i,y_i)-f(t_i,\hat{y}_i)\big)+\tfrac{h^2}{2}\,y''(\xi_i).
$$
Tomando valores absolutos y la condición de Lipschitz con constante $L$,
$$
|y_{i+1}-\hat{y}_{i+1}|\;\le\;(1+hL)\,|y_i-\hat{y}_i|+\tfrac{h^2}{2}\,M.
$$
Definiendo $a_i:=|y_i-\hat{y}_i|$, $s:=hL$ y $t:=\tfrac{h^2}{2}M$ obtenemos
$$
a_{i+1}\;\le\;(1+s)\,a_i+t,\qquad a_0=0.
$$
Aplicando la **Proposición 2**,
$$
a_{i+1}\;\le\;e^{(i+1)s}\!\left(0+\frac{t}{s}\right)-\frac{t}{s}
=\frac{t}{s}\,\big(e^{(i+1)s}-1\big)
=\frac{hM}{2L}\,\big(e^{L(t_{i+1}-a)}-1\big).
$$
Como $t_{i+1}-a=(i+1)h$, el resultado vale para $i=0,\dots,N-1$ y por tanto para todos los $i$. 

$\square$


> **Observación** La hipótesis $|y''|\le M$ puede verificarse sin conocer $y$ si $f$ es diferenciable: por regla de la cadena,
$$
y''(t)=\frac{d}{dt}f(t,y(t))=\frac{\partial f}{\partial t}(t,y(t))+\frac{\partial f}{\partial y}(t,y(t))\,f(t,y(t)),
$$
lo que permite acotar $|y''|$ a partir de cotas de $\partial_t f$, $\partial_y f$ y $f$ en el dominio de interés.

>Observación alternativa

Sea $f:\mathbb{R}^2\to\mathbb{R}$ de clase $C^1$ y define la curva
$$
\sigma(t)=(t,\,y(t)),\qquad \sigma'(t)=\big(1,\,y'(t)\big)=\big(1,\,f(t,y(t))\big).
$$
Aplicando la regla de la cadena a $f\circ\sigma$ en forma vectorial,
$$
\frac{d}{dt}\,f\big(\sigma(t)\big)\;=\;\nabla f\big(\sigma(t)\big)\cdot\sigma'(t),
$$
donde $\nabla f=(\partial_t f,\,\partial_y f)$ y “$\cdot$” denota el producto punto euclidiano. Como $y'(t)=f(t,y(t))$, resulta
$$
y''(t)\;=\;\frac{d}{dt}\,f(t,y(t))\;=\;\nabla f(t,y(t))\cdot\big(1,\,f(t,y(t))\big)
\;=\;\partial_t f(t,y(t))\;+\;\partial_y f(t,y(t))\,f(t,y(t)). 
$$

En consecuencia, si en un dominio $D$ se tienen cotas
$$
|\partial_t f|\le A,\qquad |\partial_y f|\le B,\qquad |f|\le C,
$$
entonces
$$
|y''(t)|\;=\;\big|\partial_t f+\partial_y f\,f\big|\;\le\;|\partial_t f|+|\partial_y f|\,|f|
\;\le\;A+BC\quad\text{en }D,
$$
y se puede tomar $M:=A+BC$ como cota uniforme para $|y''|$ en el Teorema.

*Alternativa (Cauchy–Schwarz):*
$$
|y''(t)|\;=\;\big|\nabla f\cdot(1,f)\big|\;\le\;\|\nabla f\|_2\,\sqrt{1+|f|^2},
$$
útil si dispones de cotas para $\|\nabla f\|_2$ y $|f|$ en $D$.


## <span style="font-family:Georgia; text-align:center;">Ejemplito</span> 
<span style="font-family:Georgia;">

Consideremos el PVI:
$$
y' = y - t^2 + 1,\qquad y(0)=0.5,\qquad 0\le t\le 2.
$$
Aquí $\partial_y f=1\Rightarrow L=1$ y $y''(t)=2-\tfrac{1}{2}e^t$, de modo que en $[0,2]$ podemos tomar $M=\max|2-\tfrac{1}{2}e^t|=\tfrac{1}{2}e^2-2$.
Usamos Euler con $h=0.2$ y comparamos $|y(t_i)-\hat{y}_i|$ con la cota del teorema.


```python
import math

def euler(f, a, b, alpha, N):
    h = (b - a) / N
    t = a
    w = alpha
    ts = [t]
    ws = [w]
    for i in range(N):
        w = w + h*f(t, w)
        t = a + (i+1)*h
        ts.append(t); ws.append(w)
    return ts, ws, h

# Problema
f = lambda t,y: y - t**2 + 1
y_exact = lambda t: (t+1)**2 - 0.5*math.e**t

a,b,alpha = 0.0, 2.0, 0.5
N = 10  # h=0.2
ts, ws, h = euler(f, a, b, alpha, N)

L = 1.0
M = 0.5*math.e**2 - 2.0
bound = lambda t: (h*M/(2*L))*(math.e**(L*(t-a)) - 1.0)

rows = []
for t, w in zip(ts, ws):
    e = abs(y_exact(t) - w)
    rows.append((t, w, y_exact(t), e, bound(t)))


import pandas as pd
df = pd.DataFrame(rows, columns=["t","ŷ (Euler)","y exacta","|error|","cota (Teorema)"])
df
```
