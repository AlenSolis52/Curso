# <span style="font-family:Georgia; text-align:center;">Método del punto medio</span>
<span style="font-family:Georgia;">

El método del punto medio está definido por definido por
$$
\hat y_0 = y_0,\qquad
\hat y_{i+1}=\hat y_i + h\, f\!\left(t_i+\frac{h}{2},\, \hat y_i+\frac{h}{2}\, f(t_i,\hat y_i)\right),\qquad i\in\{0,\ldots,N\},
$$


## <span style="font-family:Georgia; text-align:center;">Orden del error de truncamiento del método del punto medio</span>
<span style="font-family:Georgia;">

El método del punto medio tiene **error de truncamiento local de orden $r=2$**.

### <span style="font-family:Georgia; text-align:center;">Demostración</span>
<span style="font-family:Georgia;">

El **error local de truncamiento** es:
$$
\tau_{i+1}
:=\frac{1}{h}\Big(y(t_{i+1})-y(t_i)\Big)
- f\!\left(t_i+\frac{h}{2},\,y(t_i)+\frac{h}{2}\,f(t_i,y(t_i))\right),
$$
donde $y$ satisface $y'(t)=f(t,y(t))$. Supondremos que $f$ es $C^{2}$ en un rectángulo que contiene la trayectoria y que $y$ es $C^{3}$ en $[t_i,t_{i+1}]$.

Aplicando **Taylor para $y$ alrededor de $t_i$ evaluado en $t_{i+1}=t_i+h$.**  
Existe $\xi_i\in(t_i,t_{i+1})$ tal que
$$
y(t_{i+1})=y(t_i)+h\,y'(t_i)+\frac{h^{2}}{2}\,y''(t_i)+\frac{h^{3}}{6}\,y^{(3)}(\xi_i).
$$
Dividiendo entre $h$ y usando $y'(t_i)=f(t_i,y(t_i))$,
$$
\frac{1}{h}\big(y(t_{i+1})-y(t_i)\big)
= f(t_i,y(t_i))+\frac{h}{2}\,y''(t_i)+\frac{h^{2}}{6}\,y^{(3)}(\xi_i).
$$
Por la regla de la cadena,
$$
y''(t_i)=\frac{\partial f}{\partial t}(t_i,y(t_i))+\frac{\partial f}{\partial y}(t_i,y(t_i))\,f(t_i,y(t_i)).
$$
entonces
$$
\frac{1}{h}\big(y(t_{i+1})-y(t_i)\big)
= f(t_i,y(t_i))+\frac{h}{2}\Big[\frac{\partial f}{\partial t}(t_i,y(t_i))+\frac{\partial f}{\partial y}(t_i,y(t_i))\,f(t_i,y(t_i))\Big]+\frac{h^{2}}{6}\,y^{(3)}(\xi_i).
$$

<span style="font-family:Georgia;">

Si a la función $f$ le aplicamos **Taylor alrededor de $(t_{i},y(t_{i}))$** evaluado en $(t_i+\frac{h}{2},\,y(t_i)+\frac{h}{2}\,f(t_i,y(t_i)))$. Existe $\bar{\eta}_i\in(t_i,t_i+\frac{h}{2})\times(y(t_i),y(t_i)+\frac{h}{2}\,f(t_i,y(t_i)))$ tal que
$$
\begin{aligned}
f\left(t_i+\frac{h}{2},\,y(t_i)+\frac{h}{2}\,f(t_i,y(t_i))\right)= &f (t_{i},y(t_{i}))\\
&+\frac{\partial f}{\partial t}f (t_{i},y(t_{i}))\left(\left(t_i+\frac{h}{2}\right)-t_{i}\right)+\frac{\partial f}{\partial y}f (t_{i},y(t_{i}))\Big(y(t_i)+\frac{h}{2}\,f(t_i,y(t_i))-y(t_{i})\Big)\\
&+ \frac{\partial^2 }{\partial t^2 } f(\bar{\eta}_i)\left(\left(t_i+\frac{h}{2}\right)-t_{i}\right)^2+ \frac{\partial^2 }{\partial t\partial y }f(\bar{\eta}_i)\left(\left(t_i+\frac{h}{2}\right)-t_{i}\right)\Big(y(t_i)+\frac{h}{2}\,f(t_i,y(t_i))-y(t_{i})\Big)\\
&+\frac{\partial^2 }{\partial y^2 } f(\bar{\eta}_i)\Big(y(t_i)+\frac{h}{2}\,f(t_i,y(t_i))-y(t_{i})\Big)^2\\
=&f (t_{i},y(t_{i}))\\
&+\frac{h}{2}\frac{\partial f}{\partial t}f (t_{i},y(t_{i}))+\frac{\partial f}{\partial y}f (t_{i},y(t_{i}))\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)\\
&+\frac{\partial^2 }{\partial t^2 } f(\bar{\eta}_i)\left(\frac{h}{2}\right)^2+\frac{\partial^2 }{\partial t\partial y }f(\bar{\eta}_i)\left(\frac{h}{2}\right)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)+\frac{\partial^2 }{\partial y^2 } f(\bar{\eta}_i)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)^2
\end{aligned}
$$

<span style="font-family:Georgia;">

Así 
$$
\begin{aligned}
\tau_{i+1}
=& f(t_i,y(t_i))+\frac{h}{2}\Big[\frac{\partial f}{\partial t}(t_i,y(t_i))+\frac{\partial f}{\partial y}(t_i,y(t_i))\,f(t_i,y(t_i))\Big]+\frac{h^{2}}{6}\,y^{(3)}(\xi_i)\\
&- \Big[f (t_{i},y(t_{i}))\\
&+\frac{h}{2}\frac{\partial f}{\partial t}f (t_{i},y(t_{i}))+\frac{\partial f}{\partial y}f (t_{i},y(t_{i}))\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)\\
&+\frac{\partial^2 }{\partial t^2 } f(\bar{\eta}_i)\left(\frac{h}{2}\right)^2+\frac{\partial^2 }{\partial t\partial y }f(\bar{\eta}_i)\left(\frac{h}{2}\right)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)+\frac{\partial^2 }{\partial y^2 } f(\bar{\eta}_i)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)^2\Big]\\
&={h^{2}}{6}\,y^{(3)}(\xi_i)-\frac{\partial^2 }{\partial t^2 } f(\bar{\eta}_i)\left(\frac{h}{2}\right)^2-\frac{\partial^2 }{\partial t\partial y }f(\bar{\eta}_i)\left(\frac{h}{2}\right)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)-\frac{\partial^2 }{\partial y^2 } f(\bar{\eta}_i)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)^2
\end{aligned}
$$

<span style="font-family:Georgia;">

Entonces 
$$
\begin{aligned}
\tau_{i+1}
&={h^{2}}{6}\,y^{(3)}(\xi_i)-\frac{\partial^2 }{\partial t^2 } f(\bar{\eta}_i)\left(\frac{h}{2}\right)^2+\frac{\partial^2 }{\partial t\partial y }f(\bar{\eta}_i)\left(\frac{h}{2}\right)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)+\frac{\partial^2 }{\partial y^2 } f(\bar{\eta}_i)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)^2\\
 &\leq |{h^{2}}{6}\,y^{(3)}(\xi_i)|+|\frac{\partial^2 }{\partial t^2 } f(\bar{\eta}_i)\left(\frac{h}{2}\right)^2|+|\frac{\partial^2 }{\partial t\partial y }f(\bar{\eta}_i)\left(\frac{h}{2}\right)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)|+|\frac{\partial^2 }{\partial y^2 } f(\bar{\eta}_i)\Big(\frac{h}{2}\,f(t_i,y(t_i))\Big)^2|\\
\end{aligned}
$$

<span style="font-family:Georgia;">

Para concluir, supongamos que en un rectángulo $R\subset\mathbb{R}^2$ que contiene la trayectoria y el punto intermedio $\bar{\eta}_i$ se tienen las acotaciones
$$
|y^{(3)}(t)|\le M_3,\qquad 
\left|\frac{\partial^2 f}{\partial t^2}(t,y)\right|\le M_{tt},\quad
\left|\frac{\partial^2 f}{\partial t\,\partial y}(t,y)\right|\le M_{ty},\quad
\left|\frac{\partial^2 f}{\partial y^2}(t,y)\right|\le M_{yy},
$$
y además (por continuidad de $f$ en $R$ y porque la solución permanece en $R$), a lo largo de la trayectoria
$$
|f(t,y(t))|\le M_f,\qquad t\in[t_i,t_{i+1}].
$$
Entonces podemos concluir que
$$
|\tau_{i+1}|
\le \left(\frac{M_3}{6}+\frac{M_{tt}}{4}+\frac{M_{ty}M_f}{4}+\frac{M_{yy}M_f^{2}}{4}\right)h^{2}.
$$
Si definimos
$$
C:=\frac{M_3}{6}+\frac{M_{tt}}{4}+\frac{M_{ty}M_f}{4}+\frac{M_{yy}M_f^{2}}{4},
$$
obtenemos
$$
\boxed{\,|\tau_{i+1}|\le C\,h^{2}\,}.
$$


# <span style="font-family:Georgia; text-align:center;">Método de Taylor de orden 2</span>
<span style="font-family:Georgia;">

Denotemos $h:=\Delta t=(b-a)/N<1$, $t_i=a+ih$, el método de Taylor de orden 2 está definido por definido por
$$
\hat y_{i+1}=\hat y_i+h\,f(t_i,\hat y_i)+\frac{h^2}{2}\,f'(t_i,\hat y_i),\qquad \hat y_0=y_0.
$$



## <span style="font-family:Georgia; text-align:center;">Orden de convergencia del método de Taylor de orden 2</span>
<span style="font-family:Georgia;">

Si consideramos el PVI:
 $$y'=f(t,y),$$ $$y(a)=y_0.$$ 

Y se cumple lo siguiente:

a) La función $f(t,y)$ satisface la condición de Lipschitz en la variable $y$ con constante $L_1$.

b) $f'(t,y):=\dfrac{d}{dt}f(t,y)$ satisface la condición de Lipschitz en la variable $y$ con constante $L_2$.

c) $\big|y^{(3)}(t)\big|\le M$ para todo $t\in[a,b]$.

d) $\Delta t < 1$.

Entonces  el método de Taylor de orden 2 satisface
$$
\,|y(t_i)-\hat y_i|\le C\,\Delta t^{\,2}\,,\qquad i=0,\dots,N,\,
$$
donde $C$ es una constante.

### <span style="font-family:Georgia; text-align:center;">Demostración</span>
<span style="font-family:Georgia;">

Denotemos $h:=\Delta t=(b-a)/N<1$, $t_i=a+ih$. Por el teorema de Taylor con residuo, existe $\xi_i\in(t_i,t_{i+1})$ tal que
$$
y(t_{i+1})=y(t_i)+h\,y'(t_i)+\frac{h^2}{2}\,y''(t_i)+\frac{h^3}{6}\,y^{(3)}(\xi_i).
$$
Como $y'=f$ y $y''=f'(t,y)=\partial_t f(t,y)+\partial_y f(t,y)\,f(t,y)$,
$$
y(t_{i+1})=y(t_i)+h\,f(t_i,y(t_i))+\frac{h^2}{2}\,f'(t_i,y(t_i))+\frac{h^3}{6}\,y^{(3)}(\xi_i).
$$

Si $e_i:=y(t_i)-\hat y_i$. 
$$
\begin{aligned}
e_{i+1}
&=e_i+h\big[f(t_i,y(t_i))-f(t_i,\hat y_i)\big]
+\frac{h^2}{2}\big[f'(t_i,y(t_i))-f'(t_i,\hat y_i)\big]
+\frac{h^3}{6}\,y^{(3)}(\xi_i).
\end{aligned}
$$

Con las condiciones de Lipschitz y $|y^{(3)}|\le M$,
$$
|e_{i+1}|
\le \Big(1+L_1 h+\tfrac{L_2}{2}h^2\Big)|e_i|+\tfrac{M}{6}h^3.
$$
Como $h<1$, se tiene $h^2\le h$, luego
$$
|e_{i+1}|\le (1+A h)\,|e_i|+K h^3,\qquad
A:=L_1+\tfrac{L_2}{2},\; K:=\tfrac{M}{6}.
$$

**Aplicando la desigualdad de Grönwall discreta.**  
Iterando y usando $e_0=0$,
$$
|e_i|\le K h^3\sum_{j=0}^{i-1}(1+A h)^j
=\frac{K h^3}{A h}\Big((1+A h)^i-1\Big)
=\frac{K h^2}{A}\Big((1+A h)^i-1\Big).
$$
Como $(1+A h)^i\le e^{A i h}=e^{A(t_i-a)}\le e^{A(b-a)}$,
$$
|e_i|\le \frac{K}{A}\big(e^{A(b-a)}-1\big)\,h^2
=:C\,\Delta t^{\,2},\qquad i=0,\dots,N.
$$

Entonces para el método de Taylor explícito de orden $2$ se cumple
$$
\boxed{\,|y(t_i)-\hat y_i|\le C\,\Delta t^{\,2}\,,\qquad i=0,\dots,N,\,}
$$
con constante
$$
C=\frac{M}{6}\cdot\frac{e^{(L_1+\tfrac{L_2}{2})(b-a)}-1}{L_1+\tfrac{L_2}{2}}
$$
