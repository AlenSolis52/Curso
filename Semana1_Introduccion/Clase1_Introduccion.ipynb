### Recordemos la fórmula general de la longitud de arco  

Supongamos una curva $y = f(x)$ definida y derivable en $[a,b]$.  
1. Toma dos puntos muy próximos $x = a$ y $x = b$ con $|a-b| = d$.  
2. En el plano sus imágenes son $(a,f(a))$ y $(b,f(b))$, cuya distancia euclídea es  

   $$
   \sqrt{|a-b|^{2} + |f(a)-f(b)|^{2}}
   = d \,
   \sqrt{\,1 + \left(\frac{f(a)-f(b)}{\,b-a\,}\right)^{2}}.
   $$  

3. Al hacer $b \to a$ (y por tanto $d \to dx$) el cociente $\dfrac{f(a)-f(b)}{b-a}$ tiende a la derivada $f'(x)$.  
4. Sumando (integrando) todas esas “pequeñas” distancias se obtiene la longitud total  

   $$
   L \;=\; \int_a^{b} \sqrt{\,1 + \bigl(f'(x)\bigr)^{2}} \; dx.
   $$  

### Si calculamos la longitud de arco de $y = \sin x$ para el medio período  ($x \in [0,\pi]$) tenemos:



  $$
  L
  \;=\;
  \int_{0}^{\pi} \sqrt{\,1 + \cos^{2}x}\; dx.
  $$  

### La integral no admite solución elemental

1. **Cambio trigonométrico.**  
   Escribe $\cos^{2}x = 1 - \sin^{2}x$ y toma $t = \sin x$ ($dt = \cos x \, dx$). Entonces  

   $$
   L
   = \int \! \sqrt{\frac{2 - t^{2}}{1 - t^{2}}}\; dt,
   $$  

   lo que revela la estructura típica de las [integrales elípticas](https://en.wikipedia.org/wiki/Elliptic_integral).

2. **Forma elíptica.**  
   Tras un re-escalamiento puede reescribirse como  

   $$
   L
   \;=\;
   \sqrt{2}\,
   \int_{0}^{\pi/2}
     \sqrt{\,1 - k^{2}\sin^{2}\theta}\; d\theta,
   \qquad k = \tfrac{1}{2},
   $$

   que coincide con la **integral elíptica incompleta de segunda especie** $E(\phi,k)$.

3. **Teorema de Liouville.**  
   Las integrales elípticas no se expresan mediante funciones elementales (sumas, productos, exponenciales, logaritmos, etc.). Por ello  

   $$
   \int \sqrt{\,1 + \cos^{2}x}\; dx
   $$  

   carece de antiderivada elemental; su valor se describe con funciones especiales o se aproxima numéricamente.



## <span style="font-family:Georgia; text-align:center;">Ejemplo 2</span> 
<span style="font-family:Georgia;">

### Ecuación lineal de primer orden  

Considera la ecuación  

$$
x'(t)\;+\;\frac{2}{1+t^{4}}\,x(t)\;=\;\sin(t), 
\qquad t\ge 0,
$$  

que está en la forma estándar $x' + p(t)\,x = q(t)$ con  

$$
p(t)=\frac{2}{1+t^{4}}, 
\qquad 
q(t)=\sin(t).
$$  

### Calculando el factor integrante tenemos:

$$
\mu(t)=\exp\!\Bigl(\int p(t)\,dt\Bigr)
      =\exp\!\Bigl(\int\frac{2\,dt}{1+t^{4}}\Bigr).
$$  

Tras factorizar $1+t^{4}=(t^{2}+\sqrt{2}\,t+1)(t^{2}-\sqrt{2}\,t+1)$ y usar fracciones parciales se obtiene  

$$
\mu(t)=\Biggl\{
       \frac{
         \exp\!\bigl[2\tan^{-1}\!\bigl(1+\sqrt2\,t\bigr)\bigr]\,
         \bigl(t^{2}+\sqrt{2}\,t+1\bigr)
       }{
         \exp\!\bigl[2\tan^{-1}\!\bigl(1-\sqrt2\,t\bigr)\bigr]\,
         \bigl(t^{2}-\sqrt{2}\,t+1\bigr)
       }
     \Biggr\}^{1/(2\sqrt{2})},
$$  

### Al multiplicar la ecuación original por $\mu(t)$ queda exacta:

$$
\frac{d}{dt}\bigl[\mu(t)\,x(t)\bigr]\;=\;\mu(t)\,\sin(t).
$$  

Integrando de $0$ a $t$ y llamando $A=g(0)\,x(0)$ se obtiene

$$
x(t)\;=\;A\,\mu(t)\;+\;g(t)
        \int_{0}^{t}\frac{\sin(s)}{\mu(s)}\,ds.
$$  


### La integral no se puede expresar con [funciones elementales](https://en.wikipedia.org/wiki/Nonelementary_integral)

Aunque  

$$
\int_{0}^{t}\frac{\sin(s)}{\mu(s)}\,ds
$$  

**existe** para todo $t\ge 0$, su antiderivada combina $\sin(s)$ con  
$\exp\!\bigl(\int 2/(1+s^{4})\,ds\bigr)$, que ya involucra arcotangentes y términos racionales no polinomiales.  

En consecuencia, la integral solo puede  

1. describirse mediante **funciones especiales** más allá del repertorio elemental, o  
2. **evaluarse numéricamente** (por ejemplo, con reglas de cuadratura o resolviendo directamente el problema de valor inicial con Runge–Kutta).

## <span style="font-family:Georgia; text-align:center;">Ejemplo 3</span> 
<span style="font-family:Georgia;">
Sea la ecuación diferencial de primer orden  

$$
y'(t)=\exp\bigl(-t\,y(t)^{4}\bigr), \qquad t\ge 0,
$$

con alguna condición inicial $y(0)=y_{0}$.

#### 1. ¿Por qué no admite solución analítica elemental?

1. **No es lineal ni separable.**  
   Para separar variables necesitaríamos escribir $y' = g(t)\,h(y)$ y luego  
   $\displaystyle \int \frac{dy}{h(y)} = \int g(t)\,dt$.  
   Aquí, en cambio, $t$ y $y$ aparecen *dentro* del mismo exponente:  
   $\exp(-t\,y^{4})$. No es posible aislar los factores dependientes solo de $t$ o solo de $y.$

2. **Cambio de variables ineficaz.**  
   Recordemos que para **invertir la relación** y escribir $t$ como función de $y$ partimos de la ecuación original  

   $$
   \frac{dy}{dt}=e^{-t\,y^{4}}.
   $$  


   Siempre que $dy/dt\neq 0$:

   $$
   \frac{dt}{dy} \;=\; \frac{1}{\displaystyle \frac{dy}{dt}}
                     \;=\; e^{\,t\,y^{4}}.
   $$  

   Ahora $t$ es la incógnita y $y$ la variable independiente; escribimos $t=t(y)$ y la ecuación queda  

   $$
   \; \dfrac{dt}{dy}=e^{\,t(y)\,y^{4}} \;
   $$  
   ecuación implícita que tampoco se integra con funciones elementales porque $t$ sigue apareciendo en ambos lados.
