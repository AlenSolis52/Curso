# Modelo: Regulación de la **apoptosis** por p53


<span style="font-family:Georgia;">

## Introducción biológica 

### Proceso biológico

**Apoptosis.** La apoptosis es una forma de muerte celular programada esencial para el desarrollo, la homeostasis y la prevención del cáncer. Se caracteriza por eventos ordenados y por su control fino mediante redes de **señalización** y **factores de transcripción**.
### Compuestos involucrados

**p53.** La proteína p53 es un **supresor tumoral** que actúa como *sensor* y *decisor* frente a estrés genotóxico (por ejemplo, **daño en el ADN**). Una vez activada, p53 induce:
- **Arresto del ciclo** (para permitir reparación).
- **Reparación del ADN** (activando genes de reparación).
- **Apoptosis** cuando el daño es excesivo o irreparable.

**MDM2.** MDM2 es una proteína que actúa como E3 ligasa de ubiquitina. Esto significa que **puede marcar a otras proteínas** (en este caso a p53) con moléculas de ubiquitina, lo que indica que deben ser **degradadas**. Su función principal es regular los niveles de p53:
- Cuando p53 se acumula, **activa la transcripción** del gen mdm2, por lo que aumenta la cantidad de proteína MDM2.
- A su vez, **MDM2 se une a p53**, bloqueando su función y promoviendo su degradación.
- Esto crea un **bucle de retroalimentación negativa** (más p53 → más MDM2 → menos p53 → menos MDM2 → vuelve a subir p53, y así sucesivamente). 

Este bucle puede generar **oscilaciones** y **pulsos** de p53 observados en células tras daño al ADN.
En condiciones normales (sin daño celular), este bucle mantiene **bajo control los niveles de p53**.
Pero cuando hay daño en el ADN, otras proteínas inhiben temporalmente a MDM2, permitiendo que p53 aumente y active genes de reparación o apoptosis.

**$D_m$ (mediador de daño/estrés).** En este modelo, $D_m$ representa un **resumen efectivo** de señales de daño/estrés. Esto significa que no es una proteína única, sino una variable agregada que representa el estado/flujo de señalización de daño en la célula.

1) **Se produce** a un ritmo basal según el nivel de daño;  
2) Es **resuelto/atenuado** por la acción de p53;  
3) **Desestabiliza/inhibe** a MDM2 (facilitando así niveles más altos de p53).



<span style="font-family:Georgia;">

## Modelo
La dinámica del modelo es la siguiente:

$$
\begin{aligned}
\frac{d[\mathrm{p53}]}{dt} &= 
\underbrace{\phi_1}_{\substack{\text{síntesis basal}\\\text{de p53}}}
\;-\;
\underbrace{K\,[\mathrm{p53}]\,[\mathrm{Mdm2}]}_{\substack{\text{degradación de p53}\\\text{mediada por Mdm2}}},\\[3mm]
\frac{d[\mathrm{Mdm2}]}{dt} &= 
\underbrace{A\,[\mathrm{p53}]}_{\substack{\text{síntesis de Mdm2}\\\text{estimulada por p53}}}
\;-\;
\underbrace{B\,[\mathrm{Mdm2}]\,[\mathrm{Dm}]}_{\substack{\text{inhibición de Mdm2}\\\text{por la señal de daño}}},\\[3mm]
\frac{d[\mathrm{Dm}]}{dt} &= 
\underbrace{\phi_2}_{\substack{\text{síntesis de la}\\\text{señal de daño}}}
\;-\;
\underbrace{C\,[\mathrm{p53}]}_{\substack{\text{reparación del daño}\\\text{por acción de p53}}}.
\end{aligned}
$$

Por simplisidad podemos reescribirlo de forma minimalista como:
$$
\begin{aligned}
\frac{dP}{dt} &= \alpha_P \;-\; k_{MP}\,M\,P, \\[4pt]
\frac{dM}{dt} &= \alpha_M\,P \;-\; k_{MD}\,M\,D, \\[4pt]
\frac{dD}{dt} &= \alpha_D \;-\; k_{DP}\,P.
\end{aligned}
$$

<span style="font-family:Georgia;">

## Modelo **adimensional**

Para que podamos hacer un análisis más sencillo haremos lo siguiente:

Definimos **escalas** positivas $P_0,M_0,D_0$ y una escala de tiempo $T$, y variables *adimensionales*:
$$
P = P_0 x,\qquad M = M_0 y,\qquad D = D_0 z,\qquad t = T\,\tau.
$$

Usamos ahora la **regla de la cadena** para cambiar la derivada temporal de las variables dimensionales a las variables adimensionales.  
Por ejemplo, para $P(t) = P_0\,x(\tau)$ tenemos:
$$
\frac{dP}{dt} = \frac{d(P_0\,x)}{dt}
= P_0\,\frac{dx}{d\tau}\,\frac{d\tau}{dt}
= P_0\,\frac{dx}{d\tau}\,\frac{1}{T},
$$
ya que $\tau = \frac{t}{T} \;\Rightarrow\; \frac{d\tau}{dt} = \frac{1}{T}$.

De forma análoga:
$$
\frac{dM}{dt} = \frac{M_0}{T}\,\frac{dy}{d\tau}, 
\qquad 
\frac{dD}{dt} = \frac{D_0}{T}\,\frac{dz}{d\tau}.
$$

Sustituyendo estas relaciones en el sistema dimensional y dividiendo por las escalas apropiadas, obtenemos:
$$
\begin{aligned}
\frac{P_0}{T}\,\frac{dx}{d\tau} &= \alpha_P - k_{MP}\,M_0\,P_0\,x\,y,\\[4pt]
\frac{M_0}{T}\,\frac{dy}{d\tau} &= \alpha_M\,P_0\,x - k_{MD}\,M_0\,D_0\,y\,z,\\[4pt]
\frac{D_0}{T}\,\frac{dz}{d\tau} &= \alpha_D - k_{DP}\,P_0\,x.
\end{aligned}
$$
Elegimos $T$ y $M_0$ para normalizar $\dot x$ a la forma $1-xy$:
$$
\frac{dx}{d\tau} = \underbrace{\frac{\alpha_P T}{P_0}}_{=1} \;-\; 
\underbrace{k_{MP} M_0 T}_{=1}\, x y.
$$
De aquí se obtiene
$$
T=\frac{P_0}{\alpha_P},\qquad M_0=\frac{1}{k_{MP}T}=\frac{\alpha_P}{k_{MP}P_0}.
$$
Para $\dot y$ imponemos que el coeficiente de $x$ sea 1 y definimos el grupo adimensional $B$ en el término de pérdida:
$$
\frac{dy}{d\tau} = 
\underbrace{\frac{\alpha_M P_0 T}{M_0}}_{=1}\,x
\;-\;
\underbrace{k_{MD} D_0 T}_{=:B}\,y z.
$$
Con las expresiones previas,
$$
\frac{\alpha_M P_0 T}{M_0}
= \frac{\alpha_M P_0\,(P_0/\alpha_P)}{\alpha_P/(k_{MP}P_0)}
= \frac{\alpha_M k_{MP} P_0^3}{\alpha_P^2}.
$$
Elegimos $P_0$ para que valga 1:
$$
\boxed{\,P_0 = \Big(\frac{\alpha_P^2}{\alpha_M k_{MP}}\Big)^{\!\!1/3}\,},
\qquad
\Rightarrow\quad
\boxed{\,T = (\alpha_P \alpha_M k_{MP})^{-1/3}\,},
\quad
\boxed{\,M_0 = \frac{(\alpha_P \alpha_M)^{1/3}}{k_{MP}^{2/3}}\,}.
$$
Para $\dot z$, definimos $D_0$ de modo que el coeficiente de $x$ sea 1 e introducimos el parámetro $\phi_2$ como el aporte relativo de daño:
$$
\frac{dz}{d\tau} = 
\underbrace{\frac{\alpha_D T}{D_0}}_{=:\phi_2}
\;-\;
\underbrace{\frac{k_{DP} P_0 T}{D_0}}_{=1}\,x
\quad\Rightarrow\quad
\boxed{\,D_0 = k_{DP} P_0 T\,},
\qquad
\boxed{\,\phi_2 = \frac{\alpha_D}{k_{DP} P_0}\,}.
$$
Por último, el parámetro adimensional
$$
\boxed{\,B = k_{MD} D_0 T = k_{MD}\,k_{DP}\,P_0\,T^2 = \frac{k_{MD}\,k_{DP}}{\alpha_M k_{MP}}\,}.
$$
Con estas elecciones, el sistema queda exactamente en la forma **adimensional**:
$$
\begin{aligned}
\dot x &= 1 - x\,y, \\
\dot y &= x - B\,y\,z, \\
\dot z &= \phi_2 - x.
\end{aligned}
$$


<span style="font-family:Georgia;">

## Punto de equilibrio e interpretación biológica

El equilibrio $(x^*,y^*,z^*)$ satisface $\dot x=\dot y=\dot z=0$:
\begin{aligned}
\phi_2 - x^* &= 0 \;\Rightarrow\; \boxed{\,x^*=\phi_2\,},\\[4pt]
1 - x^* y^* &= 0 \;\Rightarrow\; \boxed{\,y^*=\frac{1}{\phi_2}\,},\\[4pt]
x^* - B\,y^* z^* &= 0 \;\Rightarrow\; \boxed{\,z^*=\frac{(x^*)^2}{B}=\frac{\phi_2^2}{B}\,}.
\end{aligned}
**Intuición biológica:**
- A mayor **carga de daño** ($\phi_2$), el estado estacionario muestra **más p53** ($x^*=\phi_2$), consistente con la activación de p53 por estrés.  
- El **equilibrio de MDM2** baja cuando sube p53 ($y^*=1/\phi_2$), reflejando que, en presencia de daño, MDM2 se ve más reprimida/desestabilizada.  
- El **nivel de señal de daño efectiva** en equilibrio escala como $\phi_2^2/B$: a igualdad de daño, incrementar $B$ (mayor “fuerza” con la que el daño neutraliza MDM2) reduce $z^*$.

<span style="font-family:Georgia;">

## Análisis cualitativo

La **matriz Jacobiana** $J$ en $(x^*,y^*,z^*)$ es
$$
J = \begin{pmatrix}
-\dfrac{1}{\phi_2} & -\phi_2 & 0 \\[6pt]
1 & -\phi_2^2 & -\dfrac{B}{\phi_2} \\[10pt]
-1 & 0 & 0
\end{pmatrix}.
$$
El **polinomio característico** es 
$$
\lambda^3 + \underbrace{\left(\phi_2^2+\frac{1}{\phi_2}\right)}_{p_1}\lambda^2
+ \underbrace{(2\phi_2)}_{p_2}\lambda + \underbrace{B}_{p_3},
$$, 
donde:
$$
\boxed{\,p_1 = \phi_2^2 + \frac{1}{\phi_2}\,},\qquad
\boxed{\,p_2 = 2\,\phi_2\,},\qquad
\boxed{\,p_3 = B\,}.
$$

### Criterio de Routh–Hurwitz (caso cúbico)
Un equilibrio es **asintóticamente estable** si y sólo si
$$
p_1>0,\quad p_2>0,\quad p_3>0,\quad\text{y}\quad p_1 p_2 > p_3.
$$
En el **borde de estabilidad** (cuando un par conjugado cruza el eje imaginario) se cumple
$$
\boxed{\,p_1 p_2 = p_3\,}.
$$

Aplicándolo aquí obtenemos el **umbral crítico** para $B$ (bifurcación de Hopf):
$$
\boxed{\,B_c(\phi_2) \;=\; p_1 p_2 \;=\; \big(\phi_2^2 + \tfrac{1}{\phi_2}\big)\,(2\phi_2)
\;=\; 2\,\big(1 + \phi_2^3\big)\,}.
$$

- Si $0 < B < B_c(\phi_2)$, el equilibrio es **estable** (oscilaciones amortiguadas).  
- Si $B > B_c(\phi_2)$, aparece un **ciclo límite estable**: es la firma de una **bifurcación de Hopf**.

### ¿Qué es una bifurcación de Hopf?
En una **Hopf** (típicamente *supercrítica* en modelos de regulación genética), un par de autovalores complejos cruza el eje imaginario y nace un **ciclo límite** de pequeña amplitud que **atrae** a las trayectorias vecinas. Biológicamente, esto corresponde a **oscilaciones sostenidas** de p53 y MDM2, como las observadas tras daño al ADN en células individuales.


```python

import numpy as np
import math
import matplotlib.pyplot as plt

# ------ Modelo adimensional ------
def f(t, state, phi2, B):
    x, y, z = state
    dx = 1.0 - x*y
    dy = x - B*y*z
    dz = phi2 - x
    return np.array([dx, dy, dz], dtype=float)

# ------ Umbral de Hopf ------
def hopf_threshold(phi2):
    # B_c = 2*(1 + phi2^3)
    return 2.0*(1.0 + phi2**3)

# ------ Integrador RK4 de paso fijo ------
def rk4(f, t_span, y0, args=(), h=0.01):
    t0, tf = t_span
    n = int(np.ceil((tf - t0)/h))
    t = np.linspace(t0, t0 + n*h, n+1)
    Y = np.zeros((n+1, len(y0)), dtype=float)
    Y[0] = np.array(y0, dtype=float)
    for i in range(n):
        ti = t[i]
        yi = Y[i]
        k1 = f(ti, yi, *args)
        k2 = f(ti + 0.5*h, yi + 0.5*h*k1, *args)
        k3 = f(ti + 0.5*h, yi + 0.5*h*k2, *args)
        k4 = f(ti + h, yi + h*k3, *args)
        Y[i+1] = yi + (h/6.0)*(k1 + 2*k2 + 2*k3 + k4)
    return t, Y
```

<span style="font-family:Georgia;">

## Simulaciones: por debajo y por encima de $B_c$

Fijamos $\phi_2$ y variamos $B$ a ambos lados del umbral $B_c(\phi_2)=2(1+\phi_2^3)$. Observa el decaimiento (amortiguación) vs. la aparición de un ciclo límite.

```python

# Parámetros y condiciones iniciales
phi2 = 1.0
Bc = hopf_threshold(phi2)

# --- DAMPED OSCILLATIONS: B < Bc ---
B = 0.8 * Bc
t, Y = rk4(f, (0.0, 200.0), y0=[0.2, 1.5, 0.1], args=(phi2, B), h=0.02)

# Series temporales
plt.figure()
plt.plot(t, Y[:,0])
plt.xlabel("tiempo (τ)")
plt.ylabel("x (p53)")
plt.title("Oscilaciones amortiguadas: p53 vs tiempo  (B < Bc)")
plt.show()

plt.figure()
plt.plot(t, Y[:,1])
plt.xlabel("tiempo (τ)")
plt.ylabel("y (MDM2)")
plt.title("Oscilaciones amortiguadas: MDM2 vs tiempo  (B < Bc)")
plt.show()

plt.figure()
plt.plot(t, Y[:,2])
plt.xlabel("tiempo (τ)")
plt.ylabel("z (D_m)")
plt.title("Oscilaciones amortiguadas: D_m vs tiempo  (B < Bc)")
plt.show()

# Retrato de fase (x vs y)
plt.figure()
plt.plot(Y[:,0], Y[:,1])
plt.xlabel("x (p53)")
plt.ylabel("y (MDM2)")
plt.title("Plano de fase: enfoque al equilibrio (B < Bc)")
plt.show()
```

```python

# --- LIMIT CYCLE: B > Bc ---
B = 1.2 * Bc
t2, Y2 = rk4(f, (0.0, 200.0), y0=[0.2, 1.5, 0.1], args=(phi2, B), h=0.02)

# Series temporales
plt.figure()
plt.plot(t2, Y2[:,0])
plt.xlabel("tiempo (τ)")
plt.ylabel("x (p53)")
plt.title("Oscilaciones sostenidas: p53 vs tiempo  (B > Bc)")
plt.show()

plt.figure()
plt.plot(t2, Y2[:,1])
plt.xlabel("tiempo (τ)")
plt.ylabel("y (MDM2)")
plt.title("Oscilaciones sostenidas: MDM2 vs tiempo  (B > Bc)")
plt.show()

plt.figure()
plt.plot(t2, Y2[:,2])
plt.xlabel("tiempo (τ)")
plt.ylabel("z (D_m)")
plt.title("Oscilaciones sostenidas: D_m vs tiempo  (B > Bc)")
plt.show()

# Retrato de fase (x vs y)
plt.figure()
plt.plot(Y2[:,0], Y2[:,1])
plt.xlabel("x (p53)")
plt.ylabel("y (MDM2)")
plt.title("Plano de fase: ciclo límite estable (B > Bc)")
plt.show()
```

## Parámetros críticos: cerca de $B_c$

Muy cerca de $B_c$ aparecen **transitorios largos** y oscilaciones de **amplitud pequeña**. Esto es típico de una Hopf *supercrítica*.

```python

phi2 = 1.0
Bc = hopf_threshold(phi2)

# muy cerca del umbral
B = 1.01 * Bc
t3, Y3 = rk4(f, (0.0, 800.0), y0=[0.2, 1.5, 0.1], args=(phi2, B), h=0.02)

plt.figure()
plt.plot(t3, Y3[:,0])
plt.xlabel("tiempo (τ)")
plt.ylabel("x (p53)")
plt.title("Cerca del umbral: p53 con transitorios largos (B ≈ Bc)")
plt.show()

plt.figure()
plt.plot(Y3[:,0], Y3[:,1])
plt.xlabel("x (p53)")
plt.ylabel("y (MDM2)")
plt.title("Plano de fase cerca del umbral (B ≈ Bc)")
plt.show()
```

## Mapa de bifurcación: $B_c(\phi_2) = 2(1+\phi_2^3)$

La curva separa la región de **equilibrio estable** ($B < B_c$) de la región con **ciclo límite** ($B > B_c$).


```python
phi2_vals = np.linspace(0.2, 2.5, 300)
Bc_vals = 2.0*(1.0 + phi2_vals**3)

plt.figure()
plt.plot(phi2_vals, Bc_vals)
plt.xlabel(r"$\phi_2$ (carga efectiva de daño)")
plt.ylabel(r"$B_c(\phi_2)$")
plt.title("Curva de umbral de Hopf: $B_c = 2(1 + \\phi_2^3)$")
plt.show()
```

## Interpretación biológica 

**Arquitectura del control.** El sistema implementa un **bucle negativo** p53→MDM2 ─┤ p53 **modulado por daño** ($D_m$). La **resolución de daño por p53** ($-x$ en $\dot z$) introduce un **retardo efectivo**: p53 sube, activa MDM2, pero simultáneamente reduce $D_m$, lo que a su vez **alivia** la represión sobre MDM2 con el tiempo. Esta **competencia de escalas temporales** es clave para las **oscilaciones**.

**Rol de $B$.** El parámetro $B = \dfrac{k_{MD}k_{DP}}{\alpha_M k_{MP}}$ agrupa:
- la **sensibilidad de MDM2** al daño ($k_{MD}$),  
- la **efectividad de p53** en resolverlo ($k_{DP}$),  
- y se normaliza por la **fuerza basal del bucle** p53↔MDM2 ($\alpha_M,k_{MP}$).  

Cuando **$B$ crece** (daño desestabiliza fuerte a MDM2 y/o p53 repara eficientemente), el bucle se **desincroniza** y el equilibrio pierde estabilidad vía Hopf, emergiendo **oscilaciones sostenidas** en p53 y MDM2.

**Lectura celular.** En términos celulares, un **ciclo límite** en p53 puede servir como un **código dinámico** (p. ej., número de pulsos, amplitud, duración) que coordina decisiones de **reparación** vs **apoptosis**. Valores **subcríticos** ($B<B_c$) implican respuesta **homeostática** (retorno al basal), mientras que **supracríticos** ($B>B_c$) sugieren estados de **estrés mantenido** con señales oscilatorias que pueden sesgar la **decisión apoptótica**.

**Predicciones cualitativas:**  
- Incrementar la carga de daño $\phi_2$ **eleva** el p53 de equilibrio y **reduce** el de MDM2; al mismo tiempo **desplaza** el umbral porque $B_c(\phi_2)$ crece con $\phi_2^3$.  
- Intervenciones que **inhiban MDM2** (p. ej., análogos de *nutlin*) aumentarán la amplitud/periodicidad de p53 en este esquema (efecto similar a aumentar $B$).  
- Mejoras en la **reparación** (aumento de $k_{DP}$) o en la **vulnerabilidad** de MDM2 al daño (aumento de $k_{MD}$) facilitan la entrada al régimen oscilatorio.

## Interpretación biológica 

### **Arquitectura del control.**  
El sistema implementa un **bucle negativo** p53→MDM2 ─┤ p53 **modulado por daño** ($D_m$).  
La **resolución de daño por p53** ($-x$ en $\dot z$) introduce un **retardo efectivo**: p53 sube, activa MDM2, pero simultáneamente reduce $D_m$, lo que a su vez **alivia** la represión sobre MDM2 con el tiempo.  
Esta **competencia de escalas temporales** es clave para las **oscilaciones**.

### *Explicación simple:*  
Imagina que p53 es un “bombero” que apaga incendios (daños) en la célula, y MDM2 es el “jefe” que lo manda a descansar. Cuando hay mucho fuego (daño), p53 trabaja mucho, apaga el fuego, pero mientras tanto MDM2 también se cansa y deja de mandarlo a descansar. Después, cuando ya hay poco fuego, MDM2 vuelve a decirle que pare. Este tira y afloja hace que el bombero trabaje y descanse una y otra vez, como una ola que sube y baja.


### **Rol de $B$.**  
El parámetro $B = \dfrac{k_{MD}k_{DP}}{\alpha_M k_{MP}}$ agrupa:
- la **sensibilidad de MDM2** al daño ($k_{MD}$),  
- la **efectividad de p53** en resolverlo ($k_{DP}$),  
- y se normaliza por la **fuerza basal del bucle** p53↔MDM2 ($\alpha_M,k_{MP}$).  

Cuando **$B$ crece** (el daño desestabiliza fuerte a MDM2 y/o p53 repara eficientemente), el bucle se **desincroniza** y el equilibrio pierde estabilidad vía Hopf, emergiendo **oscilaciones sostenidas** en p53 y MDM2.

### *Explicación simple:*  
$B$ es como una perilla que controla cuán fuerte reacciona la célula al daño.  
Si $B$ es pequeño, todo se calma rápido: p53 trabaja un poco y todo vuelve a la normalidad.  
Si $B$ es grande, la célula no puede encontrar su equilibrio y p53 empieza a subir y bajar una y otra vez, como un péndulo que nunca se detiene.


### **Lectura celular.**  
En términos celulares, un **ciclo límite** en p53 puede servir como un **código dinámico** (por ejemplo, número de pulsos, amplitud, duración) que coordina decisiones de **reparación** vs **apoptosis**.  
Valores **subcríticos** ($B<B_c$) implican respuesta **homeostática** (retorno al basal), mientras que **supracríticos** ($B>B_c$) sugieren estados de **estrés mantenido** con señales oscilatorias que pueden sesgar la **decisión apoptótica**.

### *Explicación simple:*  
Cuando p53 hace olas (sube y baja), la célula “piensa” que el daño sigue ahí y sigue decidiendo si debe repararse o autodestruirse (apoptosis).  
Si no hay olas y p53 se queda tranquilo, la célula entiende que todo está bien y vuelve a trabajar normalmente.


### **Predicciones cualitativas:**  
- Incrementar la carga de daño $\phi_2$ **eleva** el p53 de equilibrio y **reduce** el de MDM2; al mismo tiempo **desplaza** el umbral porque $B_c(\phi_2)$ crece con $\phi_2^3$.  
- Intervenciones que **inhiban MDM2** (p. ej., análogos de *nutlin*) aumentarán la amplitud/periodicidad de p53 en este esquema (efecto similar a aumentar $B$).  
- Mejoras en la **reparación** (aumento de $k_{DP}$) o en la **vulnerabilidad** de MDM2 al daño (aumento de $k_{MD}$) facilitan la entrada al régimen oscilatorio.

### *Explicación simple:*  
Si la célula sufre más daño, el “bombero” p53 trabaja más tiempo y el “jefe” MDM2 descansa más.  
Si ayudamos al bombero o dificultamos al jefe, el sistema entra en modo “de emergencia” y las olas de p53 se vuelven más grandes y seguidas.


### Resumen simplificado

- **Caso del punto estable ($B < B_c$):**  
  La célula recibe daño, p53 sube para repararlo y luego vuelve a su nivel normal. Todo se calma y la célula **sobrevive**.

- **Caso del ciclo límite ($B > B_c$):**  
  El daño hace que p53 y MDM2 suban y bajen sin parar. La célula no puede estabilizarse, se mantiene en “modo alarma” y puede decidir **autodestruirse (apoptosis)** para proteger al resto del organismo.
